```python
from sklearn.model_selection import train_test_split
```

```python
train_val_df,test_df=train_test_split(cloth_price_df,test_size=0.2,random_state=42)
train_df,val_df=train_test_split(train_val_df,test_size=0.25,random_state=42)
```

```python
test_df
```

```python
train_df
```

```python
val_df
```

```python
print('train_df.shape:',train_df.shape)
print('val_df.shape:',val_df.shape)
print('test_df.shape:',test_df.shape)
```

```python
input_cols=list(cloth_price_df.columns)[:-1]
target_col='Price'
```

```python
input_cols,target_cols
```

```python
train_inputs=train_df[input_cols].copy()
train_targets=train_df[target_col].copy()
```

```python
train_inputs
```

```python
train_targets
```

```python
val_inputs=val_df[input_cols].copy()
val_targets=val_df[target_col].copy()
test_inputs=test_df[input_cols].copy()
test_targets=test_df[target_col].copy()
```

```python
val_inputs
```

```python
val_targets
```

```python
categorical_cols=train_inputs.select_dtypes('object').columns.tolist()
```

```python
categorical_cols
```

```python
train_inputs[categorical_cols].describe()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder=OneHotEncoder(sparse=False,handle_unknown='ignore')
```

```python
encoder.fit(cloth_price_df[categorical_cols])
```

```python
encoder.categories_
```

```python
encoded_cols=list(encoder.get_feature_names_out(categorical_cols))
print(encoded_cols)
```

```python
train_inputs[encoded_cols]=encoder.transform(train_inputs[categorical_cols])
val_inputs[encoded_cols]=encoder.transform(val_inputs[categorical_cols])
test_inputs[encoded_cols]=encoder.transform(test_inputs[categorical_cols])
```

```python
test_inputs
```

```python
train_inputs
```

```python
from sklearn.linear_model import LinearRegression
```

```python
def rmse(targets,predictions):
    return np.sqrt(np.mean(np.square(targets-predictions)))
```

```python
X_train=train_inputs[encoded_cols]
X_val=val_inputs[encoded_cols]
X_test=test_inputs[encoded_cols]
```

```python
# Create and train the model
model=LinearRegression().fit(X_train,train_targets)
# Generate Predictions
predictions=model.predict(X_train)
#compute Loss
loss=rmse(train_targets,predictions)
print('Loss:',loss)
```

The above uses LinearRegression from Scikit learn with a loss of 52.69

```python
from sklearn.linear_model import SGDRegressor
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
```

```python
reg = make_pipeline(StandardScaler(),SGDRegressor(max_iter=1000, tol=1e-3))
```

```python
reg.fit(train_inputs[encoded_cols],train_targets)
```

```python
Newprediction=reg.predict(train_inputs[encoded_cols])
```

```python
loss=rmse(train_targets,Newprediction)
print('Loss:',loss)
```
